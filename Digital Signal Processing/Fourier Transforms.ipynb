```python
def dftmtx(N):
    n = arange(N)
    k = n.T
    U = matrix(exp(1j*2*pi/N*k*n[:,None]))
    return U
```

####Moving from a time domain to frequency domain

```python
x = hstack((ones(64), zeros(64)))
```

```python
stem(range(128),x)
```

```python
N = 128
W128 = dftmtx(N)
X = W128*matrix(x).T
```

```python
magnitude = abs(X)
stem(arange(1,129,dtype=float)/N,magnitude)
```

```python
phase = angle(X)
```

```python
stem(arange(1,129,dtype=float)/N, phase, linefmt='b-')
```

####Moving back to the time domain

```python
xx = real(W128.conj()*X/128)
stem(arange(1,129),xx)
```
