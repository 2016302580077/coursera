Suppose we would like to transmit this image over a communication channel. At the sender side, the image is projected in the canonical basis and the resulting coefficients are sent. This simply correspond to sending individually the intensity of each pixel. Suppose there is an error during transmission such that only the first half of the coefficients is correctly transmitted. The received images is only an approximation of
the original one and the missing pixels are replaced by a 0 correspond to holes in the image (in this case the entire right half) 

```python
im_approx = copy(im)
im_approx[:,32:] = 0
```

```python
imshow(im_approx)
```

```python
error = im - im_approx
distance = sqrt(multiply(error,error).sum())
distance
```

As before we project the image onto this basis and send the coefficients over a communication channel that looses the second half of the coefficients. We will first define the haar function.

```python
# This function is not implemented correctly. Need to fix
def haar(N):
    # Assume N is power of 2
    h = zeros((N,N));
    h[0,0:N] = ones((1,N))/sqrt(N);

    for k in range(1,N):
        p = fix(log(k)/log(2));
        q = k-(2**p);
        k1 = 2**p;
        t1 = N/k1;
        k2 = 2**(p+1);
        t2 = N/k2;
        for i in range(1,fix(t2)):
            h[k,i+q*t1] = (2**(p/2))/sqrt(N);
            h[k,i+q*t1+t2] = -(2**(p/2))/sqrt(N);
            
    return h
```

```python
I = matrix(im.copy().reshape(64*64,1))
shape(I)
```

```python
H = matrix(haar(4096))
shape(H)
```

```python
I_haar = H*I
shape(I_haar)
```

```python
I_haar[2049:4096] = 0;
I_haar = H.T*I_haar
I_haar.shape
```

```python
I_haar = I_haar.reshape(64,64)
imshow(I_haar)
```

```python
error = im - I_haar
distance = sqrt(multiply(error,error).sum())
distance
```
